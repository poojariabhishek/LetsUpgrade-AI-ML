# Bank Loan Modeling [Logistic Regression]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import statsmodels.api as sm
```

```python
da=pd.read_excel("data/Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
da.head()
```

```python
da.shape
```

```python
da.isnull().sum()
```

```python
da.hist(figsize=(15,12))
plt.show()
```

```python
da2=da.drop(["ID","ZIP Code"],axis=1)
da2.head()
```

```python
da2.describe().T
```

```python
x=da2.drop(["Personal Loan"],axis=1)
x.head()
```

```python
y=y = da2.iloc[:, 7:8]
y.head()
```

```python
x1 = sm.add_constant(x)
logistic = sm.Logit(y,x1)
result = logistic.fit()
```

```python
result.summary()
```

__Inference__

 Income, Family, Education, CD Account, Online and Credit Card are Most Significant as there p value is 0.<br><br>
 CCAvg and Securities Account has p value less than 0.05 so they are Significantly Important.<br>	
 Age, Experience, Mortgage has p value more than 0.05 so they are Not Significantly Important.<br>
