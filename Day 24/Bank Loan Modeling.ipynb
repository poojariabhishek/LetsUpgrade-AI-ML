# Bank Loan Modeling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
```

```python
df = pd.read_excel('data/Bank_Personal_Loan_Modelling.xlsx', sheet_name=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df = df.drop(['ID','ZIP Code'], axis=1)
df.head()
```

```python
x = df.drop(['Personal Loan'], axis=1)
x.head()
```

```python
y = df.iloc[:, 7:8]
y.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
```

```python
rf.fit(x,y)
```

```python
rf.oob_score_
```

```python
a = ['Age', 'Experience', 'Income','Family', 'CCAvg','Education', 'Mortgage', 'Securities Account','CD Account', 
     'Online', 'CreditCard']
for feature, imp in zip(a, rf.feature_importances_):
    print(feature," : ",imp);
```

__Inference:__

We can obsevrve that Age, CCAvg, Education, Family, CD Account, Mortage seem to be important features

```python
imp = df[['Income','CCAvg','Education','Family']]
imp.head()
```

```python
y.head()
```

```python
model = tree.DecisionTreeClassifier(max_depth=6)
model.fit(imp,y)
```

```python
with open('dt1.dot','w') as f:
    f = tree.export_graphviz(model, feature_names=['Income','CCAvg','Education','Family'], out_file=f);
```

```python
model.score(imp,y)
```

__Inference__:-

With score of 98.66% we can say that Income, CCAvg, Education, Family , CD Account features play important role in deciding whether the person will get a loan or not.
