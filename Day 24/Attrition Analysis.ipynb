# Attrition Analysis

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn import tree
from sklearn import preprocessing
```

```python
df = pd.read_csv('data/general_data.csv')
df.head()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(11.28) 

df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(2.6)
```

```python
df = df.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = preprocessing.LabelEncoder()
```

```python
df['Attrition'] = le.fit_transform(df['Attrition'])
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['Department'] = le.fit_transform(df['Department'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['Gender'] = le.fit_transform(df['Gender'])
df['EducationField'] = le.fit_transform(df['EducationField'])
df.head()
```

```python
x = df.drop(['Attrition'], axis=1)
x.head()
```

```python
y = df.iloc[:,1:2]
y.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
```

```python
rf.fit(x,y)
```

```python
rf.oob_score_
```

```python
a = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome','Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked','PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion','YearsWithCurrManager']
for feature,imp in zip(a, rf.feature_importances_):
    print(feature," : ",imp);
```

__Inference__

Looking at the importance score for each feature we can conclude that features like Age, MonthlyIncome, TotalWorkingYears, DistanceFromHome, PercentSalaryHike, YearsAtCompany seems to be most importance among other features.

```python
imp = df[['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany']]
imp.head()
```

```python
model = tree.DecisionTreeClassifier(max_depth=12)
model.fit(imp,y)
```

```python
with open('dt2.dot','w') as f:
    f = tree.export_graphviz(model, feature_names=['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany'], out_file=f);
```

```python
model.score(imp,y)
```

__Inference__

With score of 94.80% we can say that Age, MonthlyIncome, TotalWorkingYears, DistanceFromHome, PercentSalaryHike, YearsAtCompany features play important role in attrition level in the company.
