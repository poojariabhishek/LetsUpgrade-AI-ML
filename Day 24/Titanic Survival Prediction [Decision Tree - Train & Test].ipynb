# Titanic Survival Prediction [Decision Tree-Train & Test]

__TRAIN DATASET__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn import preprocessing
```

```python
data = pd.read_csv('data/train.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.drop(['Cabin','PassengerId','Name','Ticket'], axis=1, inplace=True)
data.head()
```

```python
label=preprocessing.LabelEncoder()
data['Sex']=label.fit_transform(data['Sex'])
data.head()
```

```python
pred = pd.DataFrame([data['Sex'],data['Age'],data['Fare']]).T
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=6)
tree_model.fit(X=pred, y=data['Survived'])
```

```python
with open('tree1.dot','w') as f :
    f = tree.export_graphviz(tree_model, feature_names=['Sex','Age','Fare'], out_file=f);
```

```python
Accuracy = tree_model.score(X=pred, y=data['Survived'])
Accuracy
```

__TEST DATASET__

```python
data_t = pd.read_csv('data/test.csv')
data_t.head()
```

```python
data_t.isnull().sum()
```

```python
data_t['Sex'] = label.fit_transform(data_t['Sex'])
data_t['Embarked'] = label.fit_transform(data_t['Embarked'])
data_t.head()
```

```python
test_features = data_t[['Age', 'Sex', 'Fare']]
test_features.head()
```

```python
test_preds = tree_model.predict(test_features)
```

```python
predicted_output = pd.DataFrame({'passengerId':data_t['PassengerId'],'Survived':test_preds})
```

```python
predicted_output.to_csv('Output.csv',index=False)
```
